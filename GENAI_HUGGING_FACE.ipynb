```python
# Step 2: Import libraries
from transformers import pipeline
import gradio as gr
```

```python
# Step 3: Load the sentiment analysis pipeline
# This automatically downloads a pre-trained model
classifier = pipeline("sentiment-analysis")
```

```python
# Step 4: Define a function for sentiment analysis
def analyze_text(text):
    result = classifier(text)[0]
    return f"Label: {result['label']}, Score: {round(result['score'], 4)}"
```

```python
# Step 5: Create the web interface
demo = gr.Interface(
    fn=analyze_text,
    inputs="text",
    outputs="text",
    title="My First AI Sentiment App",
    description="Type a sentence below to see if it's Positive or Negative!"
)
```

```python
# Step 6: Launch the app
demo.launch(share=True)
```

```python
from transformers import pipeline
import gradio as gr
```

```python
# Load summarization model
summarizer = pipeline("summarization")
```

```python
def summarize_text(text):
    summary = summarizer(text, max_length=60, min_length=20, do_sample=False)
    return summary[0]['summary_text']
```

```python
# Web app
app = gr.Interface(
    fn=summarize_text,
    inputs="text",
    outputs="text",
    title="Hugging Face Text Summarizer",
    description="Paste a long paragraph to get a short summary"
)
```

```python
app.launch(share=True)
```
